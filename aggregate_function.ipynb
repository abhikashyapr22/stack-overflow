```python
import pandas as pd
```

```python
data = {
    'Id':[1,1,1,1,1],
    'date':['12/02/2022','13/02/2022','14/02/2022','15/02/2022','16/02/2022'],
    'Steps':[12,0, 45,0,67]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df[df['Id']==1].groupby('Id').agg(Sum_Tot_Steps=('Steps','sum'))
```

```python
df2 = pd.DataFrame(df.groupby('Id').sum().reset_index(drop=True), columns=['Id','TotalSteps','Zero_steps'])
```

```python
df2
```

```python
df2['Id'] = df.groupby('Id')['Steps'].sum().index
```

```python
df2
```

```python
Zero_steps =  df[df['Steps']==0].groupby('Id')['Steps'].count().reset_index(drop=True)
Zero_steps
```

```python
Total_steps = df.groupby('Id')['Steps'].sum().reset_index(drop=True)
Total_steps
```

```python
ID = df.groupby('Id')['Steps'].sum().index
ID
```

```python
data1 = {
    'Id':ID,
    'Total_steps':Total_steps,
    'Zero_steps':Zero_steps
}
```

```python
df2 = pd.DataFrame(data1)
```
